**See if there is a correlation between columns** When we are building a linear model, we want to look for strong correlations between columns were trying to predict and columsn were using to predict that value, which allows a linear model to make good predictions

Turn the values into numeric numbers just in case but you can check data type first before having to do so by doing the .dtypes 

```python
cols_to_convert = ['wins', 'losses', 'points', 'total_yards']

# Convert columns to numeric, coercing errors to NaN
for col in cols_to_convert:
    NFLteams[col] = pd.to_numeric(NFLteams[col], errors='coerce')

# Check for NA values in these columns
na_values = NFLteams[cols_to_convert].isna().sum()
print("NA values after conversion:", na_values)

# Drop rows with any NA values in these columns
NFLteams.dropna(subset=cols_to_convert, inplace=True)
```

**Lets take the last 2 years 2022 and 2023 and put them in our test data set and put the rest in our training data set. We want to train our model on the training set and see how the model is doing on the test set. We want the model to be evaluated on data it hasn't seen before**

```python
train = NFLteams[NFLteams["year"] < 2019].copy()
test = NFLteams[NFLteams["year"] >= 2019].copy()
```

Lets look at the shape (how many rows and columns) aim for something close to 80/20 SPLIT of train and test

```python
train.shape
```

```python
test.shape
```

**Lets being training our model**

```python
from sklearn.linear_model import LinearRegression

reg = LinearRegression()
```

Define predictors (columns you are going ot use, to predict your target)

```python
predictors = ["points"]
target = ["total_yards"]
```

Call the fit method to fit your linear regression model 
Pass in columns to train (predictors) and the target 
After you fit your model, you can use your regression to make predictions

```python
reg.fit(train[predictors], train["total_yards"])
```

We pass predict method and only pass in predictors for test, because we want it to create predictions without know what the actual answers are

```python
predictions = reg.predict(test[predictors]) 
```

Adding predictions to our NFLteams data frame

```python
print(len(NFLteams))
print(len(predictions))
```

```python
# This is assuming that your test set is based on the years 2019 and beyond
test_indices = NFLteams.loc[NFLteams["year"] >= 2019].index

# Now assign predictions to the 'predictions' column only for the test set
NFLteams.loc[test_indices, 'predictions'] = predictions
NFLteams
```

```python
NFLteams.tail(5)
```

```python
NFLteams.columns
```

```python
from sklearn.metrics import mean_absolute_error

error = mean_absolute_error(test["total_yards"], predictions)
error
```

**^What that means is that we were within 251 total yards based on the mont of points a team scored for that season**

```python
NFLteams.describe()["total_yards"] #You want to make sure your error is less than the standard deviation
```

We can Look team by team to see how our predictions look

```python
test['predictions'] = predictions # Added the predictions to the 'test' DataFrame 
test[test["team"]=="Tennessee Titans"]
```

To get the difference of how many total yards there were vs how many we predicted 

```python
errors = (test["total_yards"] - test["predictions"]).abs()
errors
```

Lets group by team - which will create a seperate group by team and find the mean error by team which gives you how many points we were off by each team

```python
error_by_team = errors.groupby(test["team"]).mean()
error_by_team
```

We can see how many yards each team had by average

```python
total_yards_by_team = test["total_yards"].groupby(test["team"]).mean()
```

This gives the error ratio which means what percent I am of the actual count

```python
error_ratio = error_by_team/total_yards_by_team
error_ratio
```

```python
error_ratio.plot.hist()
```

Shows teams where my error is lowest to highest, closest to accuracy
